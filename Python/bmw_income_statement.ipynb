```python
import numpy as np
import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
file_name = 'Financial-statements-2021'
column_name = 'links'
df = pd.read_csv('/content/sample_data/bmwgroup_url.csv')

# Display the DataFrame
df.columns
df_links = df['links']
year =2021
df_timeseries = pd.DataFrame()
def initial_df():
  print('*************************')
  global df_timeseries
  df_timeseries = pd.DataFrame(df.iloc[0]).T


  # Set the index of 'df_timeseries' to 2021

  df_timeseries.index = [2021]
  df_timeseries = df_timeseries.rename_axis('Year')

for link in df_links:
  print(link)
  df = pd.read_excel(link)
  #Remove first three lines
  # Find the index of the text you want to use as a delimiter
  delimiter = "/20"

  # Get the substring before the delimiter
  get_year = link[link.find(delimiter):]
  get_year = int(get_year[1:5])
  print(get_year)
  if get_year > 2016:
    df = df.drop([0,1,2])

    # Take away odd columns with spaces (nans)
    df = df.drop(columns=df.columns[1::2])

    df.set_index(df.columns[0], inplace=True)

    df.dropna()

    #Remove previous year data
    df = df.drop(columns=df.columns[1::2])

  else:
    df = df.drop([0,1,2,3,4,5,6])

   # Take away odd columns with spaces (nans)

    #df = df.drop(columns=df.columns[1::2])

    df.set_index(df.columns[0], inplace=True)

    df = df.dropna(how='all')

    df = df.dropna(axis=1,how='all')


  df = df.T
  print(df.head())
  if len(df_timeseries) < 1:
    initial_df()
  df_timeseries.loc[year] = df.iloc[0]

  year -=1
print(df_timeseries.index)

df_timeseries.to_csv('/content/sample_data/bmw_years.csv')
```
